```python
import pandas as pd
```

```python
#1 questão
try:
    real_estate_evaluation_data = pd.read_excel('real_estate_evaluation.xlsx')
except:
    print("Erro ao carregar arquivo")
finally:
    print("Arquivo carregado com sucesso!")
```

```python
try:
    correlation = real_estate_evaluation_data.corr()

    column = 'X2 house age'
    corr_column = correlation[column].drop(index=column).sort_values(ascending=False)


    two_columns = corr_column[:2].index.tolist()

    print(corr_column)
    print('\n', two_columns)
except:
    print("Erro ao correlacionar ou selecionar variáveis")
finally:
    print("Correlação concluída com sucesso!")
```

```python
#2 questão
import matplotlib.pyplot as plt
```

```python
try:
    plt.figure(figsize=(10, 6))
    plt.scatter(real_estate_evaluation_data['X5 latitude'], real_estate_evaluation_data['X4 number of convenience stores'], alpha=0.7, color='pink')
    plt.title('X5 latitude X X4 number of convenience stores')
    plt.xlabel('X4 number of convenience stores')
    plt.ylabel('X5 latitude')
    plt.grid(True)
    plt.show()
except:
    print("Não foi possível gerar o gráfico")
finally:
    print("Gráfico gerado com sucesso!")
```

```python
#3 questão
import traceback

try:
    wine_data = pd.read_excel('wine.xlsx')
except:
    print("Erro ao carregar arquivo")
finally:
    print("Arquivo carregado com sucesso!")
```

```python
try:
    print("Colunas: ", wine_data.columns)

    class_column = 'Target'
    statistics_study = wine_data.groupby(class_column).describe()
    print("Estudo Estatístico por Classe:", statistics_study)
except KeyError as e:
    print("Erro: Coluna de classe não encontrada. Verifique o nome da coluna.")
    print(traceback.format_exc())
except Exception as e:
    print("Erro ao realizar estudo estatístico:")
    print(traceback.format_exc())
finally:
    print("Processo concluído com sucesso!")
```

```python
x = 'Alcohol'
y = 'Malic acid'

classes = wine_data['Target'].unique()
colors = ['pink', 'blue', 'orange']

plt.figure(figsize=(10, 6))

for i, class_value in enumerate(classes):
    class_data = wine_data[wine_data['Target'] == class_value]
    plt.scatter(class_data[x], class_data[y], color=colors[i], label=f'Classe {class_value}')

plt.title(f'Gráfico de Dispersão: {x} X {y}')
plt.xlabel(x)
plt.ylabel(y)
plt.legend(title="Classes")
plt.show()

```
